### `if` Koşullu İfadesi

* Python’da koşullu mantığı ifade etmenin yolu if ifadeleridir.
* Hemen hemen her programlama dili bu kavrama sahiptir.
* If ifadeleri, bir koşul tanımlar ve bu koşul True (doğru) olduğunda çalıştırılacak bir dizi komutu belirtir.

Bir prototip...

```
if some_expression:    
    do_this()
    do_that()
```

Eğer `if` (koşul) doğruysa, ***girintili*** ifadeler (komutlar) yürütülür. Aksi takdirde, bu girintili ifadeler atlanır ve program yürütülmeye `if` ifadesinden sonraki satırdan devam eder.



`````
month = "jan"
summer_months = ["may", "jun", "jul", "aug"]

if month in summer_months:
    print("yay - it's still summer!")
else:
    print("boo - have to go to school today!")
print("program done")
`````

### `else` İfadesi

Bazen `if` koşuluna bir alternatif belirtmek isteriz; bunu `else`  ifadesiyle yaparız.

Örneğin…

```
if <condition>:
    <block1>
else:
    <block2>
```

* Eğer koşul doğru ise block1 çalıştıırlacaktır.
* Eğer koşul yanlış ise block1 çalıştıırlacaktır.

**Else ifadesi, Python’da “aksi halde…” demenin bir yoludur.**

Nasıl ki `if`  blokları girinti ile tanımlanıyorsa,  `else`  blokları da girintiyle tanımlanır.

Örneğin şöyle:

```
    <statement1>

    <statement2>
    <statement3>
```

```

    <statement1>
else:
    <statement2>

```

### `elif` İfadesi

Bazen if ve else bölümleri arasında bir veya daha fazla ara koşul tanımlamamız gerekir.

Örneğin:
`if A then do X, else if B then do Y, otherwise do Z`

Bunu Python’da ifade etmek için `elif` ifadesini kullanırız.

```
if condition1:
    do_thing_1()
elif condition2:
    do_thing_2()
else:
    do_thing_3()
```
* Eğer `condition1` doğruyla, `do_thing_1()` kod parçacığı çalıştırılır.
* Aksi durumda, eğer `condition2` doğruyla, `do_thing_2()` kod parçacığı çalıştırılır.
* Hiçbiri doğru değilse, `do_thing_3()` kod parçacığı çalıştırılır.


* `elif` blokları, tıpkı `if` ve `else` blokları gibi girinti (indentation) kullanılarak tanımlanır.

* Her önemli koşul için ayrı bir if/elif kullanmak iyidir; hataları ilgilendiğin durumlarla karıştırmamak gerekir.

Örneğin, yalnızca 1 ve 2 değerleriyle ilgileniyorsan ve diğer tüm değerleri hata olarak değerlendiriyorsan, şu kodu yazabilirsin:

```
if month == "jan":      # deal with jan here
    apply_jan_discount()
elif month == "feb":    # deal with feb here
    apply_feb_discount()
else:                   # deal with errors here
    apply_no_discount()
```
bu daha iyi:
```
if month == "jan":
    apply_jan_discount()
else:    # x must be 2 then, right? not necessarily!
    apply_feb_discount()
```
İkinci (sonraki) kod, geçerli bir durumu hata durumlarıyla birleştirerek hataları gizler.


#### Alıştırma

if/elif/else ifadesi için aşağıdaki kodu inceleyin...
```
print("1")
if condition1:    
    print("2")
elif condition2:    
    print("3")
else:    
    print("4")
print("5")
```

Aşağıdaki durumlarda çıkış ne olacaktır..
* her iki koşul (condition) False olursa?  
* her iki koşul (condition) True olursa?   
* `condition1`, `True` ve `condition2` ise `False` olursa?   
* `condition1`, `False` ve `condition2` ise `True` olursa?

Tahminini test etmek için, aşağıdaki hücreyi `condition1` ve `condition2` değişkenlerini True ve False değerlerinin dört olası kombinasyonuyla çalıştırarak dene.

`````
condition1 = False
condition2 = True

print("1")
if condition1:
    print("2")
elif condition2:
    print("3")
else:
    print("4")
print("5")
`````

### İç İçe `if` İfadeleri (Nested `if` Statements)


* Bazen bir if ifadesini başka bir if ifadesinin içine yerleştirmemiz gerekir.
* Bunlara iç içe (nested) if ifadeleri denir.
* if ifadelerini, bir `if` bloğunun, `elif` bloğunun veya `else` bloğunun herhangi bir kısmına yerleştirebiliriz.
* `if` ifadelerini istediğimiz kadar derin biçimde iç içe yerleştirebiliriz. Örneğin şöyle olabilir...

```
if condition1:
    if condition2:
        if condition3:
        ...
```


#### Alıştırma

```
print("1")
if condition1:    
    print("2")    
    if condition2:        
        print("3")    
    else:        
        print("4")
else:    
    print("5")
print("6")
```
Aşağıdaki durumlarda çıkış ne olacaktır..
* her iki koşul (condition) False olursa?  
* her iki koşul (condition) True olursa?   
* `condition1`, `True` ve `condition2` ise `False` olursa?   
* `condition1`, `False` ve `condition2` ise `True` olursa?

Tahminini test etmek için, aşağıdaki hücreyi `condition1` ve `condition2` değişkenlerini True ve False değerlerinin dört olası kombinasyonuyla çalıştırarak dene.

`````
condition1 = False
condition2 = True

print("1")
if condition1:
    print("2")
    if condition2:
        print("3")
    else:
        print("4")
else:
    print("5")
print("6")
`````
