### `while` Döngüleri

* `while` ifadesi, Python’da tekrarlanabilir mantığı ifade etmenin bir yoludur.
* `while` ifadeleri bir koşul ve girintili komutlardan oluşan bir blok tanımlar; biçimi şöyledir:
```
while condition:    
    block of indented statements
    ...
print('done with loop')
```

* Koşul döngüye girerken ve döngü bloğu her çalıştırıldıktan sonra kontrol edilir.
* `if` ifadesine benzer, ancak koşul doğru olduğu sürece blok tekrar tekrar çalıştırılır.
* Koşul yanlış olduğunda döngü sonlanır; yani tekrar etme durur ve kontrol döngüden sonraki ifadeye geçer.
* Eğer koşul başlangıçta yanlışsa, döngüye hiç girilmez; yani döngü bloğu bir kez bile çalıştırılmaz.

#### Döngü Sayacı (Loop Counter)
* döngü gövdesine girişini kontrol eden değişkendir
* çoğu zaman döngünün kaç kez çalıştığını takip eder
* başlatılması (initialize edilmesi) gerekir
* döngünün istenen sayıda çalışmasını sağlamak için döngü bloğu içinde bir noktada değiştirilmesi gerekir


#### Atama Operatörleri (Assignment Operators)
- `i = i + 1` ifadesi şu şekilde yazılabilir:  `i += 1`
- Daha genel olarak, `i = i + x` ifadesi `i += x` şeklinde yazılabilir; bu, x değerini i'ye ekler ve sonucu tekrar `i` değişkenine atar.
- Bu sözdizimini döngüler içinde sık sık görürsünüz; örneğin, döngü sayacını her yinelemede artırmak istediğimizde.
- Diğer aritmetik operatörler için de benzer atama operatörlerimiz vardır:


| operator | equivalent |
| -------- | ---------- |
| x += y   | x = x + y  |
| x -= y   | x = x - y  |
| x *= y   | x = x * y  |
| x /= y   | x = x / y  |
| x **= y  | x = x ** y |
| x %= y   | x = x % y  |

### `continue` İfadesi

* continue , döngünün en üstüne anında atlar.
* Döngü koşulu yeniden test edilir ve koşul doğruysa bir sonraki yineleme başlar; yanlışsa döngü sonlanır.
* Python’da bunun anlamı şudur: “baştan alalım.”

#### `continue` Örnek

* `start` ve `end` değerlerini pozitif varsayalım
* Eğer `start` değeri geçersizse, `end` değerini sormaya gerek yoktur.
```
start = 0
end = 0
while (start <= 0) or (end <= 0):    
    if start <= 0:
        start = int(input('starting integer (>0): '))        
        if (start <= 0):            
            continue
    if end <= 0:
        end = int(input('ending integer (>0): '))
        if end <= 0:            
            continue
    print(‘using start’, start, ‘and end’, end)
```

````
start = 0
end = 0
while (start <= 0) or (end <= 0):
  if start <= 0:
      start = int(input('starting integer (>0): '))
      if (start <= 0):
          continue
  if end <= 0:
      end = int(input('ending integer (>0): '))
      if end <= 0:
          continue
  print(f"using start {start} and end {end}")
````

### `for` Döngüleri

* `while`  döngüleri, bir koşula bağlı olarak program mantığını tekrarlar.
* `for` döngüleri ise bir dizinin (sequence) içeriklerine bağlı olarak program mantığını tekrarlar.
* Bu döngülere, dizinin elemanları üzerinde iterasyon yapar (iterate over a sequence) deriz.

Örneğin, aşağıdaki program bir string içindeki karakterleri, her satıra bir karakter gelecek şekilde yazdırır:
```
mystr = 'this is a test'
for i in mystr:
    print(i)
```
Sonuç:
```

i
s
...
```


````
for i in 'Python':
    print(i)
````

#### Sayısal `for`  Döngüsü

* For döngüleri, tıpkı while döngülerine benzer şekilde, bir sayı dizisi ile de kullanılabilir.
* Örneğin, aşağıdaki program ilk üç tam sayıyı ve karelerini yazdırır:
```

    print(i, i*i)
```

```

```
* While döngülerine göre büyük bir avantajına dikkat edin: döngü sayaç değişkenini ayrıca başlatmaya veya artırmaya gerek yoktur.
* Ancak sayı aralığını tek tek yazmak zahmetlidir.

````
for i in (1,2,3):
  print(i, i*i)
````

#### `range()` Fonksiyonu

* Python’daki `range()`  fonksiyonu bir sayı dizisi üretir.
* Kullanım biçimi (prototype): range([start], end, [increment])
* start isteğe bağlıdır ve varsayılan değeri 0’dır.
* end zorunludur ve diziyi end-1 değerinde sonlandırır.
* increment dizideki elemanlar arasındaki artış miktarını belirler.
* increment isteğe bağlıdır ve varsayılan değeri 1’dir.
* `range()` fonksiyonu, Python’da "range" adı verilen özel bir tür döndürür.

````
for i in range(0, 10):
  print(i, i*i)
````

````
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
````

### Soru 2

Hogwarts’ta gece yarısı dolaşan Harry, Sırlar Odası’na açılan gizli bir kapı keşfeder.
* Kapının üzerinde bir yazı belirir: “Bu kapı, doğru sihirli sayıyı söyleyene açılacaktır.”
* Kapının büyüsü şu şekilde çalışır:
  - Kapı, 1 ile 60 arasında rastgele bir sihirli sayı belirler.
  - Harry her denemede bir sayı söyler.
  - Eğer Harry’nin tahmini:
    - Sihirli sayıdan büyükse:
      - Kapı “Alohomora yetersiz… Daha küçük bir sayı söyle!” diye fısıldar.
    - Sihirli sayıdan küçükse:
      - Kapı “✨ Lumos yetmedi… Daha büyük bir sayı söyle!” der.
  - Harry doğru sayıyı bulana kadar kapı sihirli ipuçları vermeye devam eder.
  - Harry’nin toplam kaç deneme yaptığı tutulmalıdır.
  - Tüm süreç while döngüsü ile yapılmalıdır.

````
import random as rnd
randomSayi = rnd.randint(1,61)
tryNumber = 0
while True:
    tryNumber += 1
    userIn = int(input("Bu kapı, doğru sihirli sayıyı söyleyene açılacaktır:"))
    if userIn >  randomSayi:
        print("Alohomora yetersiz… Daha küçük bir sayı söyle!")
    elif userIn < randomSayi:
        print("✨ Lumos yetmedi… Daha büyük bir sayı söyle!")
    else:
        print("✨✨✨✨Kapı açıldı.✨✨✨✨")
        break

print(f"Doğru Sayı: {randomSayi}")
print(f"Deneme Sayısı: {tryNumber}")
````
