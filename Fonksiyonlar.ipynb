#### DDocstring’ler

### Örnek Fonksiyon

````
# N'in faktöriyeli 1*2*...*N şeklinde tanımlanır.
# İşte 10'un faktöriyelini yazdıran örnek bir fonksiyon tanımı.
def fact10():
    '''
    10'un faktöriyelini yazdır.
    10'un faktöriyeli 1*2*...*10 olarak tanımlanır.
    '''
    limit = 10
    result = 1
    for i in range(1, limit+1):
        result *= i # result = result * i
    print(result)
````

````
# Bu fonksiyon hakkında yardım al
help(fact10)
````

````
# Burada da fonksiyonu çağıralım
fact10()
````

### Bir Fonksiyona Değer Aktarma

* `fact10()`oldukça sınırlıdır.
* Sadece tek bir değerin (10'un) faktöriyelini yazdırır.
* Bunun yerine, herhangi bir sayının faktöriyelini yazdırabilecek daha esnek bir sürüm tanımlayabilsek güzel olurdu, örneğin şöyle…

````
def fact(num):
    '''Herhangi bir sayının faktoriyelini hesaplar'''
    result = 1
    for i in range(1, num+1):
        result *= i
    print(result)

fact(20)
````

#### Parametreler ve Argümanlar

Bir fonksiyon içinde, çağıran tarafından gönderilen değerleri almak için tanımladığımız değişkenlere parametre denir.

* Bu kodda `a`, `b` ve `c` birer parametredir:
```
def sum(a, b, c):
    return a + b + c
```
* Bir fonksiyon çağrılırken çağıran tarafından sağlanan değerlere ise argüman denir.
* Bu kodda `1`, `2`, ve `3`argümandır:

```
sum(1, 2, 3)
```


#### Argüman Aktarma

* Fonksiyonlar sıfır dahil istediğimiz sayıda parametre tanımlayabilir.


### Dönen Değerler (Return Values)

Sonucu ekrana yazdırmak yerine, fonksiyonun sonucu geri döndürmesini sağlayabiliriz. Böylece çağıran kod bu sonucu yazdırabilir veya bir hesaplamada kullanabilir.

```
def fact(num):
    ''' herhangi bir sayının faktöriyelini hesapla '''
    fact = 1
    for i in range(1, num+1):
        fact *= i
    return fact
```
Bu fonksiyonu şu şekilde çağırırız:
```
f = fact(5)
print('5! = ', f)
```

* Fonksiyonlar, çağırana değeri `return` ifadesiyle geri döndürür.
* `return` ifadesi iki şeyin gerçekleşmesine neden olur:
  - fonksiyon sona erer ve program yürütmesi, fonksiyonun çağrıldığı yerden hemen sonra devam eder
  - döndürülen değer çağırana iletilir
* İstediğiniz kadar return ifadesi kullanabilirsiniz (sıfır tane dahil).
* Çağıran kod dönen değeri kullanmak istiyorsa, onu bir değişkene kaydetmeli veya bir ifadede kullanmalıdır...
```
resp = input('Yes or No (y/n)? ')
```
* Eğer çağıran dönen değeri yok sayarsa, değer kaybolur...
```
input('Yes or No (y/n)? ') # this loses the answer
```


#### Birden Fazla Değer Döndürme

* Fonksiyonlar birden fazla değer döndürebilir.
* `return`  ifadesinde virgülle ayırarak birden fazla değer belirtmeniz yeterlidir.

```
def divide(dividend, divisor):
    quotient = int(dividend / divisor)
    remainder = dividend % divisor
    return quotient, remainder
```

* Dönen değerleri şu şekilde alabilirsiniz:

```
(q, r) = divide(100, 9)
print(f'quotient={q}, remainder={r}')
```


````
def divide(dividend, divisor):
  quotient = int(dividend / divisor)
  remainder = dividend % divisor
  return quotient, remainder

(q, r) = divide(100, 9)
print(f"quotient={q}, remainder={r}")
````

## İsim Uzayları (Namespaces)

* Bir `namespace`  (isim uzayı), belirli bir bağlamda var olan soyut bir değişken koleksiyonudur.
  - Bir şirketteki personel ID numaraları gibi düşünebilirsiniz.
  - Benim ID numaram sadece çalıştığım şirketin kapsamı içinde geçerlidir.
* Bu koddaki bir isim uzayı örneğini daha önce görmüştük:
  
```
import random
x = random.randrange(10)
```

* Bu durumda, random modülünün programınızın geri kalanından tamamen ayrı kendi isim uzayı vardır ve bu isim uzayına nokta (.) gösterimi ile erişilir.
* Eğer random. önekini kullanmazsanız, bu kez bambaşka bir isim uzayına — programınızın global isim uzayına — başvurmuş olursunuz.

### Global İsim Uzayı

* Her Python kaynak dosyasının kendi global isim uzayı vardır.
* Global isim uzayı, Python dosyasında herhangi bir fonksiyonun dışında tanımlanan tüm adları (değişkenler ve fonksiyonlar) içerir.
* Aşağıdaki kodda `count`, `mystr` ve `req` değişkenlerinin tamamı global değişkendir; yani hepsi global isim uzayında bulunur:

```
count = 0
while True:
    mystr = input('? ')
    if (mystr == '1'):
        req = 'add'
    else:
        break
    count += 1
```


### Yerel (Local) İsim Uzayı

* Bir fonksiyon tanımladığınızda, Python o fonksiyon için bir yerel isim uzayı oluşturur; bu uzay fonksiyon gövdesi içinde oluşturulan tüm parametreleri ve değişkenleri kapsar.
* Yerel isim uzayındaki değişkenler, global isim uzayındaki değişkenlerden ayrı ve bağımsızdır.
* Bir fonksiyon içinde, aynı isimli global bir değişken varsa, yerel değişken global değişkenin yerini alır (önceliklidir).
* Yerel değişkenler geçicidir — yalnızca fonksiyon çalıştığı süre boyunca var olurlar.

### Yerel bir değişkenin global bir değişkenin yerine geçtiğini gösteren örnek

````
var = 1
def myfunc():
    var = 2  # local scope
    print('local var:', var)
myfunc()
print('global var:', var)
````

### Yerel isim uzayının geçici (transient) doğasını gösteren örnek

````
def func():
    var2 = 2  # local scope
    print('local var:', var2)
func()
print('global var:', var2)
````

## Modüller (Modules)

* Bir modül, yeniden kullanılabilir Python kodu içeren bir dosyadır.
* Modülün adı, .py uzantısı olmadan dosya adıdır.
* Modüllerin nasıl içe aktarılacağını (import) zaten gördük:

```
import random
```
* Python’ın sizin modüllerinizi görebilmesi için, modül dosyasının (.py uzantılı) çalıştırdığınız programla aynı klasörde veya sistem tarafından tanınan özel bir konumda olması gerekir.
  - Python içindeki `sys.path`
  - Shell ortamındaki `PYTHONPATH`

### Modül İsim Uzayı

* Her modülün, ana kaynak dosyasının global isim uzayından bağımsız kendi isim uzayı vardır.

* Bir modülün içindeki nesnelere şu genel söz dizimiyle erişilebilir:
```
module-name.variable
module-name.function(<args>)
```
* Örneğin, daha önce gördüğümüz gibi:

```
rand_val = random.randrange(0, 10)
```


````
import random
print(random.random())
print(random.randint(1, 10))
````

###  `from` Deyimi

* Kod içe aktarmak için şu söz dizimini de kullanabilirsiniz:

```
from module-name import *
```
* Bu ifade, belirtilen modüldeki tüm adları (*) global isim uzayına yükler.
* Bu tür bir import işleminde, modüldeki isimler global isim uzayına geldiği için erişim yaparken `module-name.`. önekini kullanmanız gerekmez.
* Örneğin şöyle yapabilirsiniz:

```
from random import *
rand_val = randrange(0, 10)
# I didn't need to use random.randrange(0, 10)
```

* Bir modülden yalnızca seçili isimleri de içe aktarabilirsiniz:

```
from random import randrange, randint
```

* Bu ifade, belirtilen modülden sadece açıkça listelenen adları (`randrange` ve `randint`) global isim uzayına yükler.
* Önceki örnekte olduğu gibi, bu import işleminden sonra adlar artık global isim uzayında bulunduğundan, erişirken önek kullanmaya gerek yoktur:

```
from random import randrange, randint
rand_val = randrange(0, 10)
```

````
from random import random, randint
print(random())
print(randint(1, 10))
````

### `import` ya da `from` Ne Zaman Kullanılır?

* Genel olarak `import` kullanmak daha iyidir çünkü…
  - isim çakışmaları ve beklenmedik durumlar daha az olur
  - kodunuzu daha açık ve anlaşılır kılar

* Bazen bir modülün fonksiyonlarını çok sık kullandığınız için `from` ile doğrudan global isim uzayına aktarmak daha pratik olabilir.
* Bu da gayet normaldir, ancak dikkatli yapılmalıdır ve isim çakışmalarına karşı dikkatli olmak gerekir.
